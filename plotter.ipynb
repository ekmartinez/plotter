```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import ticker
plt.style.use('seaborn-whitegrid')
%matplotlib inline
```

```python
data = [23, 21, 14, 31]
labels = ['0-35', '31-63', '61-95', '97+']
explode = (0, 0, 0, 0.05)

fig1, ax1 = plt.subplots()
ax1.pie(data, explode=explode,labels=labels, autopct='%1.1f%%',
        startangle=90)
ax1.axis('equal')

plt.title('Data Example')
plt.legend(loc='best')
plt.show()
```

```python
data = {'Bush': 315000, 
        'Brady': 400600, 
        'Burress': 105250, 
        'Thompson': 145890,
        'Miller': 156541}

customers = list(data.keys())
values = list(data.values())

fig = plt.figure(figsize = (10, 5))
ax = plt.axes()
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))

plt.bar(customers, values, color='maroon', width= 0.9)

plt.xlabel('Players')
plt.ylabel("Salaries")

plt.title('Top five')
#plt.rcParams['figure.dpi'] = 300
plt.show()
```
